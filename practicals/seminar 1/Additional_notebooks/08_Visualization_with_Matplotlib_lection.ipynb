```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Отрисовка графика "по умолчанию":

```python
# создаем синтетические данные
import numpy as np
x = np.linspace(0,15,100)
# рисуем график по умолчанию
plt.plot(x)
```

```python
# график по умолчанию 2
y = np.cos(x)
plt.plot(y)
```

```python
# график по умолчанию 3
y = np.cos(x)
plt.plot(x, y)
```

Чем плох график выше?

* отсутствуют подписи осей
* отсутствует название графика

Давайте это поправим:

```python
plt.plot(x, y)
plt.title('cos(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
```

Добавим лейблы на график:

```python
z = 0.5 * np.sin(x)

# рисуем график
plt.plot(x, y, label='cos(x)')
plt.plot(x, z, label='sin(x)')
plt.title('0.5 sin(x) и cos(x)')
plt.xlabel('х')
plt.ylabel('y, z')
plt.legend()
plt.show()
```

Все равно что-то не так... Что? 

Давайте поменяем размер!

```python
# рисуем график
plt.figure(figsize=(8,4))
plt.plot(x, y, label='cos(x)')
plt.plot(x, z, label='sin(x)')
plt.title('0.5 sin(x) и cos(x)')
plt.xlabel('х')
plt.ylabel('y, z')
plt.legend()
plt.show()
```

Как поменять цвет?

```python
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, y, label='cos(x)', c='g')
plt.plot(x, z, label='sin(x)', c='r')
plt.title('0.5 sin(x) и cos(x)')
plt.xlabel('х')
plt.ylabel('y, z')
plt.legend()
plt.show()
```

Параметр linestyle:

```python
styles = ['-', '--', '-.', ':', 'steps']

for ls in styles:
    # рисуем график
    plt.figure(figsize=(15,6))
    plt.plot(x, z, c='r', ls=ls)
    plt.title('0.5 sin(x)')
    plt.xlabel('х')
    plt.ylabel('y')
    plt.show()
```

Параметр marker:

```python
styles = ['+', ',', '.', '1', '2', '3', '4' ]

for marker in styles:
    # рисуем график
    plt.figure(figsize=(15,6))
    plt.plot(x, z, c='r', marker=marker, ls=':', mec='b')
    plt.title('0.5 sin(x)')
    plt.xlabel('х')
    plt.ylabel('y')
    plt.show()
```

Контроль толщины линии графика осуществляется за счет параметра linewidth:

```python
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, z, c='cornflowerblue', linewidth=2)
plt.title('0.5 sin(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
```

## Полярные координаты

```python
r = np.arange(0, 2, 0.01)

theta = 2 * np.pi * r

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # less radial ticks
ax.set_rlabel_position(-22.5)  # get radial labels away from plotted line
ax.grid(True)

ax.set_title("График линии в полярных координатах a.k.a. Cпираль Архимеда")
plt.show()
```

## Контурные графики

Подробнее с примерамми кода:

https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.contour.html

$f(x, y) = x^2 + y^2$

$f(x, y) = const$

```python
def f(x,y):
    return x**2 + y**2
```

```python
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.figure(figsize=(5,5))
plt.contour(X, Y, Z)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.contourf(X, Y, Z)
plt.show()
```

### Преобразования изображений

```python
import matplotlib.transforms as mtransforms

def do_plot(ax, Z, transform):
    im = ax.imshow(Z, interpolation='none',
                   origin='lower',
                   extent=[-2, 4, -3, 2], clip_on=True)

    trans_data = transform + ax.transData
    im.set_transform(trans_data)

    # display intended extent of the image
    x1, x2, y1, y2 = im.get_extent()
    ax.plot([x1, x2, x2, x1, x1], [y1, y1, y2, y2, y1], "y--",
            transform=trans_data)
    ax.set_xlim(-5, 5)
    ax.set_ylim(-4, 4)


# prepare image and figure
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
Z = face.copy()

# image rotation
do_plot(ax1, Z, mtransforms.Affine2D().rotate_deg(30))

# image skew
do_plot(ax2, Z, mtransforms.Affine2D().skew_deg(30, 15))

# scale and reflection
do_plot(ax3, Z, mtransforms.Affine2D().scale(-1, .5))

# everything and a translation
do_plot(ax4, Z, mtransforms.Affine2D().
        rotate_deg(30).skew_deg(30, 15).scale(-1, .5).translate(.5, -1))

plt.show()
```
